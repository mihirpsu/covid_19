```python
import pandas as pd
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv")
```

```python
df
```

```python
dates=list(set(df['date']))
dates.sort()
```

```python
for date in dates:
    newdf=pd.DataFrame(columns=['State','County','Confirmed'])
    newdf['State']=df[df['date']==date]['state'].values
    newdf['County']=df[df['date']==date]['county'].values
    newdf['Confirmed']=df[df['date']==date]['cases'].values
    newdf['FIPS']=df[df['date']==date]['fips'].values
    t=date.split("-")
    newdf.to_csv("CTY_level_COVID19_{}_{}_{}.csv".format(t[0],int(t[1]),int(t[2])),index=False)
    print(date,df[df['date']==date]['cases'].sum())
```
